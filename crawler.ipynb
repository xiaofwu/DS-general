**Crawl Rotten Tomato webpage for best-movies-2022**

```python
import requests
import lxml
from bs4 import BeautifulSoup

url = "https://editorial.rottentomatoes.com/guide/best-movies-2022/"
headers = {
  'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 13_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
}

f = requests.get(url, headers = headers)
webpage = f.text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
soup
```

```python
movies_list = soup.find_all('h2')
movies_list
```

**Get the movie names**

```python
movies = []
for movie in movies_list:
    temp = movie.find('a')
    if temp:
        movies.append(temp.text)
movies
```

```python
movies.remove('Related News')
movies.remove('More Countdown')
```

```python
movies
```

```python
movies_url = []
for movie in movies_list:
    temp = movie.find('a')
    if temp:
        movies_url.append(temp['href'])
movies_url
```

```python
movies_url.remove('https://editorial.rottentomatoes.com/more-related-content/?ref=214601')
movies_url.remove('https://editorial.rottentomatoes.com/countdown/')
```

```python
#get the movie introductions
from random import random
import time
movie_content_web = []
for url in movies_url:
    t = 1 + 2 * random()
    time.sleep(t)
    movie_f = requests.get(url, headers = headers)
    movie_soup = BeautifulSoup(movie_f.content, 'lxml')
    movie_content_web.append(movie_f)
```

```python
movie_content = []
for movie_f in movie_content_web:    
    movie_soup = BeautifulSoup(movie_f.content, 'lxml')
    temp = movie_soup.find("div", {"id": "movieSynopsis"})
    movie_content.append(temp.text)
```

```python
movie_critics = []

for movie_f in movie_content_web:    
    movie_soup = BeautifulSoup(movie_f.content, 'lxml')
    temp = movie_soup.find(attrs={'data-qa': 'critics-consensus'})
    if temp:
        movie_critics.append(temp.get_text())
    else:
        movie_critics.append("")
```

```python
movie_critics
```

```python
#write to excel file
import lxml
from xlwt import *

workbook = Workbook(encoding = 'utf-8')
table = workbook.add_sheet('data')
table.write(0, 0, 'Number')
table.write(0, 1, 'movie_url')
table.write(0, 2, 'movie_name')
table.write(0, 3, 'movie_introduction')
table.write(0, 4, 'movie_critics_consensus')

print(len(movies))
print(len(movies_url))
print(len(movie_content))
print(len(movie_critics))
```

```python
for line in range(1, len(movies)+1):
    table.write(line, 0, line)
    table.write(line, 1, movies_url[line-1])
    table.write(line, 2, movies[line-1].strip())
    table.write(line, 3, movie_content[line-1].strip())
    table.write(line, 4, movie_critics[line-1].strip())
    
workbook.save('2022_best_movies.xls')
```
